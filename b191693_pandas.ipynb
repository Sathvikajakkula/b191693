```python
import pandas as pd
```

```python
df = pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv")
```

```python
df.head(3)
```

```python
df.tail(5)
```

```python
#no of roes and columns
print(df.shape)
```

```python
df=df.dropna()#remove problematic rows
```

```python
print(df.shape)
```

```python
rooms=df["Rooms"]
price=list(df["Price"])
print(rooms)
```

```python
import matplotlib.pyplot as plt
```

```python
print(max(price)-min(price)/10)
fig=plt.figure()
plt.hist(x=price, edgecolor ="black",bins=1000)
plt.title("histogram")
plt.xlabel("price range")
plt.ylabel("no of houses")
plt.show()
```

```python
rooms_count = df["Rooms"].value_counts()
rooms_count = dict(rooms_count)
print(rooms_count)
room_size =rooms_count.keys()
no_of_rooms = rooms_count.values()
fig = plt.figure(figsize=(6,6))
plt.pie(x = no_of_rooms,labels = room_size)
plt.legend()
plt.show()
```

```python
region_count= df["Regionname"].value_counts()
region_count = dict(region_count)
#print(region_count)
a =region_count.keys()
b = region_count.values()
fig = plt.figure(figsize=(5,5))
plt.pie(x = b,labels = a)
plt.legend()
plt.show()
```

```python
type_count= df["Type"].value_counts()
type_count = dict(type_count)
#print(region_count)
c=list(type_count.keys())
d= list(type_count.values())
fig = plt.figure(figsize=(5,5))
plt.bar(x = c,height = d)
plt.show()
```

```python
price= df["Price"]
price = list(price)
distance=df["Distance"]
distance=list(distance)
fig = plt.figure(figsize=(50,50))
plt.plot(distance,price,marker="o",linestyle="",color="r")
plt.show()
```
